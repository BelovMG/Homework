```python
import numpy as np
import matplotlib.pyplot as plt
l = 5
x = np.arange(0, 1, 0.01)
ehist = -np.log((np.random.uniform(0, 1, 1000)))/l
plt.plot(x, l * np.exp(-l * x))
plt.hist(ehist, bins = 100, density = True)
plt.show()
```

```python
from matplotlib.patches import Circle
fi = np.linspace(0, 2 * np.pi, 100)
plt.plot(np.sin(fi), np.cos(fi))
p = np.random.uniform(0, 2 * np.pi, 500)
r = np.sqrt(np.random.uniform(0, 1, 500))
plt.scatter(np.cos(p) * r, np.sin(p) * r)
plt.axis('equal')
plt.show()
```
